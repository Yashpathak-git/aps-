# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n = len(graph)
    dp = [[-1]*(1<<n) for _ in range(n)]
    path_tracker = [[-1]*(1<<n) for _ in range(n)]

    def visit(city, mask):
        if mask == (1<<n) - 1:
            return graph[city][0]

        if dp[city][mask] != -1:
            return dp[city][mask]

        ans = float('inf')
        best_next_city = -1

        for next_city_candidate in range(n):

            if not (mask & (1<<next_city_candidate)):

                cost_to_next = graph[city][next_city_candidate]
                remaining_cost = visit(next_city_candidate, mask | (1<<next_city_candidate))
                new_cost = cost_to_next + remaining_cost

                if new_cost < ans:
                    ans = new_cost
                    best_next_city = next_city_candidate

        dp[city][mask] = ans
        path_tracker[city][mask] = best_next_city
        return ans

    min_cost = visit(0, 1)

    path = []
    current_city = 0
    current_mask = 1

    while True:
        path.append(current_city)
        if current_mask == (1 << n) - 1:
            break

        next_city = path_tracker[current_city][current_mask]
        if next_city == -1:
            break

        current_city = next_city
        current_mask |= (1 << current_city)

    path.append(0)

    return min_cost, path
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

min_cost, path = tsp(graph)
print("Minimum travelling cost:", min_cost)
print("Optimal path:", path)
```
